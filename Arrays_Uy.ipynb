```python
from ast import Index
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")
def reverse(arr):
    reversed_arr = arr[::-1]
    print(reversed_arr)

#sort()
def sort(arr):
  print("""
  1. Increasing Order
  2. Decreasing Order
  """)
  choice = int(input("Order: "))
  if choice == 1:
    arr.sort()
  elif choice == 2:
    arr.sort(reverse=True)
#insert()
def insert(arr):
  index = int(input("Index: "))
  add = int(input("Insert: "))
  arr.insert(index, add)
  print(arr)

#delete()
def delete(arr):
  index = int(input("Index: "))
  del arr[index]

#modify()
def modify(arr):
  index = int(input("Index: "))
  edit = int(input("New: "))
  arr[index] = edit

def main():
    # Initialize an array
    arr = [10, 60, 30, 40, 50]
    print("Original array:", arr)
    print()

    while True:
        print("Current Listing:")
        for i, element in enumerate(arr):
          print(f"Index {i}: {element}")

        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Sort")
        print("4. Insert")
        print("5. Delete")
        print("6. Modify")
        print("7. Exit")
        print()

        choice = int(input("Enter your choice (1 to 7): "))

        if choice == 1:
          traverse_array(arr)
        elif choice == 2:
          reverse(arr)
        elif choice == 3:
          sort(arr)
        elif choice == 4:
          insert(arr)
        elif choice == 5:
          delete(arr)
        elif choice == 6:
          modify(arr)
        else:
          print()
          print(f"Final: {arr}")
          break
        print()

if __name__ == "__main__":
    main()
```
